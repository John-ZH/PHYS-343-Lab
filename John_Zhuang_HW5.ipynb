```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
plt.rcParams["figure.figsize"] = (15,10)
```

# Problem setup

```python
#Use a scale of 3
bkgd = stats.norm.rvs(loc=0, scale=3, size=1000000, random_state=None)
plt.hist(bkgd, density=True)
```

```python
sgnl = np.random.uniform(0,20,1000000)
plt.hist(sgnl, density=True)
```

```python
data = bkgd + sgnl

signaledges = np.linspace(0,20,40)
dataedges = np.linspace(-7,27,68)

Psd, temp, temp2= np.histogram2d(data,sgnl, bins=[dataedges,signaledges], density=True)

datacenters = (dataedges[:-1] + dataedges[1:]) / 2
signalcenters = (signaledges[:-1] + signaledges[1:]) / 2
plt.pcolormesh(datacenters,signalcenters,Psd.T)
plt.ylabel('True signal, $P(s|d)$', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
```

# Problem 2

```python
bkgd = stats.norm.rvs(loc=6, scale=3, size=1000000, random_state=None)
sgnl = np.random.uniform(0,20,1000000)
data = bkgd + sgnl

signaledges = np.linspace(0,20,40)
dataedges = np.linspace(-1,33,68)

Psd, temp, temp2= np.histogram2d(data,sgnl, bins=[dataedges,signaledges], density=True)

datacenters = (dataedges[:-1] + dataedges[1:]) / 2
signalcenters = (signaledges[:-1] + signaledges[1:]) / 2
plt.pcolormesh(datacenters,signalcenters,Psd.T)
plt.ylabel('True signal, $P(s|d)$', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
```

```python
plt.suptitle('Probability distribution of the observed data for true signal of 6.15', fontsize=30)
plt.xlabel('Observed data', fontsize = 24)
plt.ylabel('Probability density', fontsize = 24)
plt.step(datacenters, Psd[:,12])
```
